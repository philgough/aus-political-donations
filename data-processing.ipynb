```python
sd['party'] = ''
for i, row in sd.iterrows():
    party = ''
    if row.coalition:
        party += 'Coalition'
        
    if row.labor:
        party += 'Labor'
        
    if row.greens:
        party += 'Greens'
    if row.katter:
        party += 'Bob Katter'
    if row.phon:
        party += 'One Nation'
    if row.ldp:
        party += 'Liberal Democrats'
    if row.fish:
        party += 'Fishers'        
    if row.lambie:
        party += 'JLN'
    if party == '':
        party = 'other'
    sd.at[i, 'party'] = party
```
